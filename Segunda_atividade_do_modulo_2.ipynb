```python
import numpy as np
import pandas as pd

#!pip install --upgrade numpy
#!pip install --upgrade pandas
#!pip install --upgrade matplotlib

df = pd.read_csv('https://pycourse.s3.amazonaws.com/bike-sharing.csv')
print(df)
```

```python
#4. Responder às seguintes perguntas:
#a. Qual o tamanho desse dataset?
size = df.size
print(size)
```

```python
#b. Qual a média da coluna windspeed?
wcondition = df['weather_condition'].describe()[['mean']]
print('A média da coluna weather_condition é\n', wcondition)
```

```python
#c. Qual a média da coluna temp?
temp = df['temp'].describe()[['mean']]
print('A média da coluna temp é\n', temp)
```

```python
#d. Quantos registros existem para o ano de 2011?
df['datetime'] = pd.to_datetime(df['datetime'])
df = df.set_index('datetime')
```

```python
df.dropna(axis='index', inplace=True)
```

```python
maxloca = df.copy()
```

```python
maxloca['datetime'] = pd.to_datetime(maxloca['datetime'])
```

```python
filtered11 = maxloca.loc[(maxloca['datetime'] >= '2011-01-01') & (maxloca['datetime'] < '2012-01-01')]
```

```python
filtered11
```

```python
sum11 = filtered11['total_count']
print("Quantidade de locacoes em 2011:", sum(sum11))
```

```python
filtered12 = maxloca.loc[(maxloca['datetime'] > '2011-12-31') & (maxloca['datetime'] <= '2012-12-31')]
```

```python
filtered12
```

```python
sum12 = filtered12['total_count']
print("Quantidades de locacoes em 2012:", sum(sum12))
```
